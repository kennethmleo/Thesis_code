```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob
%matplotlib inline
```

```python
filenames = glob.glob('*.csv')
```

```python
df_brownian = pd.DataFrame()
display(df_brownian)

df_brownian_not = pd.DataFrame()
display(df_brownian_not)
```

```python
counter = 1
brownian_X = []
brownian_Y = []
for i in filenames:
    dataframe = pd.read_csv(i)
    df_brownian['brownian' + str(counter)] = dataframe['brownian']
    df_brownian = df_brownian.fillna(0)
    x = np.arange(0,201,1)
    y = np.array(df_brownian['brownian' + str(counter)])
    for j in x: brownian_X.append(j)
    for k in y: brownian_Y.append(k)
    counter +=1
    

display(df_brownian.tail())

counter_not = 1
not_brownian_X = []
not_brownian_Y = []
for i in filenames:
    dataframe_not = pd.read_csv(i)
    df_brownian_not['not_brownian' + str(counter_not)] = dataframe_not['not_brownian']
    df_brownian_not = df_brownian_not.fillna(0)
    x = np.arange(0,201,1)
    y = np.array(df_brownian_not['not_brownian' + str(counter_not)])
    for j in x: not_brownian_X.append(j)
    for k in y: not_brownian_Y.append(k)
    counter_not +=1
    
display(df_brownian_not.tail())
print(len(df_brownian_not))
```

```python
nodes = np.arange(0,200,1)

plt.figure(figsize=(10*2,7))

plt.subplot(121)
for column in df_brownian:
    plt.plot(nodes, df_brownian[column], color='grey', linewidth=0.5)

plt.axhline(y=0, color = 'black')
plt.xlabel('Number of nodes disconnected', fontsize=20)
plt.ylabel('Assortativity', fontsize=20)
#plt.xticks(np.arange(0,1.2,0.2), fontsize=15)
plt.yticks(np.arange(0,1.2,0.2), fontsize=15)

plt.subplot(122)
for column in df_brownian_not:
    plt.plot(nodes, df_brownian_not[column], color='grey', linewidth=0.5)

plt.axhline(y=0, color = 'black')
plt.xlabel('Number of nodes disconnected', fontsize=20)
plt.ylabel('Assortativity', fontsize=20)
#plt.xticks(np.arange(0,1.2,0.2), fontsize=15)
plt.yticks(np.arange(0,1.2,0.2), fontsize=15)

#plt.savefig('Assortativity_100trials.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,7))
for column in df_brownian:
    plt.plot(nodes, df_brownian[column], color='grey', linewidth=0.5)

plt.axhline(y=0, color = 'black')
plt.xlabel('Number of nodes disconnected', fontsize=20)
plt.ylabel('Assortativity', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(np.arange(0,1.2,0.2), fontsize=15)

plt.savefig('nodes_100.png', bbox_inches='tight')
plt.show()
```

```python
df_brownian['75th'], df_brownian['25th'] = np.percentile(df_brownian, 75, axis=1), np.percentile(df_brownian, 25, axis=1)
df_brownian_not['75th'], df_brownian_not['25th'] = np.percentile(df_brownian_not, 75, axis = 1), np.percentile(df_brownian_not, 25, axis = 1)
```

```python
df_brownian['median'] = df_brownian.median(axis=1)
display(df_brownian.head())

df_brownian_not['median'] = df_brownian_not.median(axis=1)
display(df_brownian_not.head())
```

```python
nodes = np.arange(0,200,1)
brownian = np.array(df_brownian['median'].tolist())
brownian_max = np.array(df_brownian['75th'].tolist())
brownian_min = np.array(df_brownian['25th'].tolist())

not_brownian = np.array(df_brownian_not['median'].tolist())
not_brownian_max = np.array(df_brownian_not['75th'].tolist())
not_brownian_min = np.array(df_brownian_not['25th'].tolist())

print(len(df_brownian))
```

```python
plt.figure(figsize=(10*2,7))
plt.subplot(1,2,1)
plt.fill_between(nodes,brownian_max, brownian_min, color="#3F5D7D")
plt.plot(nodes,brownian, linewidth = 2, color='white', label = 'Brownian')
plt.xlabel('Number of nodes disconnected', fontsize=25)
plt.ylabel('Assortativity', fontsize=25)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
#plt.legend(fontsize=20)
plt.ylim(0,1)

plt.subplot(1,2,2)
plt.fill_between(nodes,not_brownian_max, not_brownian_min, color="#FF8C00")
plt.plot(nodes,not_brownian, linewidth = 2, color='white', label = 'Non-Brownian')
plt.xlabel('Number of nodes disconnected', fontsize=25)
plt.ylabel('Assortativity', fontsize=25)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
#plt.legend(fontsize=20)
plt.ylim(0,1)

plt.show()
```
